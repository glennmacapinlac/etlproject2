```python
#SOCCER TABLES
Match_Odd = "Resources/Matches_Odds.csv"
Odd_df = pd.read_csv(Match_Odd)
Odd_df.head()
```

```python
Match_Result = "Resources/Matches_Results.csv"
Result_df = pd.read_csv(Match_Result)
Result_df.head()
```

```python
Top_Book = "Resources/top5bookies.csv"
Top_df = pd.read_csv(Top_Book)
Top_df.head()
```

```python
# Create a filtered dataframe from specific columns
odd_cols = ["match_id", "home_team_name", "away_team_name","home_team_odd","away_team_odd","tie_odd"]
odd_transform= Odd_df[odd_cols].copy()

# Rename the column headers
odd_transform = odd_transform.rename(columns={"match_id": "ID", 
                                              "home_team_name":"Home_Name", 
                                              "away_team_name":"Away_Name",
                                              "home_team_odd":"Home_Odd",
                                              "away_team_odd":"Away_Odd",
                                              "tie_odd":"Tie_Odd"})

# Clean the data by dropping duplicates and setting the index
odd_transform.drop_duplicates("ID", inplace=True)
odd_transform.set_index("ID", inplace=True)

odd_transform.head()
```

```python
# Create a filtered dataframe from specific columns
result_cols = ["match_id","home_team_name","away_team_name","home_team_score","away_team_score", "final_result"]
result_transform = Result_df[result_cols].copy()

# Rename the column headers
result_transform = result_transform.rename(columns={"match_id": "ID", 
                                              "home_team_name":"Home_Name", 
                                              "away_team_name":"Away_Name",
                                            "home_team_score":"Home_Score",
                                            "away_team_score":"Away_Score",
                                            "final_result":"Final_Result"})

# Clean the data by dropping duplicates and setting the index
result_transform.drop_duplicates("ID", inplace=True)
result_transform.set_index("ID", inplace=True)

result_transform.head()
```

```python
# Create a filtered dataframe from specific columns
top_cols = ["ID","Visitor_Team","Home_Team","Visitor_Odd","Draw_Odd","Home_Odd","Bookies_choice","Visi_Team_PTS", "Home_Team_PTS", 
            "True_Result", "Winner_Odd"]
top_transform = Top_df[top_cols].copy()

# Clean the data by dropping duplicates and setting the index
top_transform.drop_duplicates("ID", inplace=True)
top_transform.set_index("ID", inplace=True)

top_transform.head()
```

```python
#UFC TABLE
ufc1_file = "Resources/data.csv"
ufc1_df = pd.read_csv(ufc1_file)
ufc1_df.head()
```

```python
ufc1_df.insert(0, 'New_ID', ufc1_df.index + 0)

ufc1_df
```

```python
ufc2_file = "Resources/ufc-master.csv"
ufc2_df = pd.read_csv(ufc2_file)
ufc2_df.head()
```

```python
ufc2_df.insert(0, 'New_ID', ufc2_df.index + 0)

ufc2_df
```

```python
# Create a filtered dataframe from specific columns
ufc1_cols = ["New_ID", "R_fighter", "B_fighter", "R_odds", "B_odds", "Winner"]
ufc1_dropcols= ufc1_df[ufc1_cols].copy()

# Rename the column headers
ufc1_dropcols = ufc1_dropcols.rename(columns={"New_ID": "Fight_ID",
                                                    "R_fighter": "Red_Corner_Fighter",
                                                    "B_fighter": "Blue_Corner_Fighter",
                                                    "R_odds": "Red_Fighter_odds",
                                                    "B_odds": "Blue_Fighter_odds",
                                                    "Winner": "Winner"})


ufc1_dropcols
```

```python
ufc2_cols = ["New_ID", "R_fighter", "B_fighter", "R_odds", "B_odds", "Winner", "R_ev", "B_ev"]
ufc2_dropcols = ufc2_df[ufc2_cols].copy()

# Rename the column headers
ufc2_dropcols = ufc2_dropcols.rename(columns={"New_ID": "Fight_ID",
                                                    "R_fighter": "Red_Corner_Fighter",
                                                    "B_fighter": "Blue_Corner_Fighter",
                                                    "R_odds": "Red_Fighter_odds",
                                                    "B_odds": "Blue_Fighter_odds",
                                                    "Winner": "Winner",
                                                    "R_ev": "Red_profit_on_a_100_credit_winnin_bet",
                                                    "B_ev": "Blue_profit_on_a_100_credit_winnin_bet"})



ufc2_dropcols
```

```python
combined_ufc_df = pd.merge(ufc1_dropcols, ufc2_dropcols, how='outer', on='Fight_ID')
combined_ufc_df
```

```python
combined_ufc_df = combined_ufc_df.rename(columns={"Red_Corner_Fighter_x":"Red_Corner_Fighter", "Blue_Corner_Fighter_x":"Blue_Corner_Fighter", "Red_Fighter_odds_x":"Red_Fighter_odds",  "Blue_Fighter_odds_x":"Blue_Fighter_odds", "Winner_x":"Winner"})

# Rename our _y columns 
combined_ufc_df = combined_ufc_df.rename(columns={"Red_Corner_Fighter_y":"Red_Corner_Fighter", "Blue_Corner_Fighter_y":"Blue_Corner_Fighter", "Red_Fighter_odds_y":"Red_Fighter_odds",  "Blue_Fighter_odds_y":"Blue_Fighter_odds", "Winner_y":"Winner"})

combined_ufc_df
```

```python
# Create the dataframe 

ufc_transformed = pd.DataFrame(combined_ufc_df) 

  
# dropping the rows having NaN values 

ufc_transformed = ufc_transformed.dropna() 

  
# To reset the indices  

ufc_transformed = ufc_transformed.reset_index(drop = True) 

  
# Print the dataframe 
ufc_transformed
```

```python
ufc_transformed.to_csv("Resources/ufc_transformed.csv",
                  encoding="utf-8", index=False, header=True)
```

```python
#Transform df by selecting just the desired columns. 
#Note that data from the 2013-2014 or older only has the opening odds 

#Add column that can summarize int Home/Away scores with Boolean Home wins instead
Odds_df['Home Win'] =  np.where(Odds_df["Home Score"] > Odds_df["Away Score"], True, False)

#Select desired columns
Opening_Odds_df = Odds_df[["Date", "Home Team", "Away Team", "Home Win", "Home Odds Open", "Away Odds Open"]]

#drop NaN columns
Opening_Odds_df.dropna

Opening_Odds_df.tail()
```
